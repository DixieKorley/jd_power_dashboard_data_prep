```python
def teb_split_rows(df_list):
    merged_pairs = [pd.merge(df_list[i], df_list[i+1], how="inner", on=["Dimension", "Measure"]) for i in range(0, len(df_list), 2)]
    merged_pairs = pd.concat(merged_pairs, axis=0)
    return merged_pairs

def transform_electric_business(df, df2, df3):
    rows = []
    for col in df3:
        temp_df = pd.DataFrame({
            "Dimension": df[2:],
            "Measure": df2[2:], 
        })
        # print(df3[col])
        if df3[col].iloc[1] == "Rank":
            temp_df["Rank"] = df3[col].iloc[1:]
        else:
            temp_df["Brand"] = df3[col].iloc[1]
            temp_df["Score"] = df3[col].iloc[1:]
        rows.append(temp_df)
    return rows
```

```python
def merge_transform_sat_kpi(df_list):
    df1 = []
    df2 = []
    merged_list = []
    for i in range(len(df_list)):
        if i % 2 == 0:
            df1.append(df_list[i])
        else:
            df2.append(df_list[i])
    
    for d1, d2 in zip(df1,df2):
        merged = pd.merge(d1, d2, on=["Question", "Met"], how='inner')
        merged_list.append(merged)
    new_df = pd.concat(merged_list, axis=0, ignore_index=True)
    new_df = new_df.rename(columns={"Brand Name_x": "Brand Name"})
    del new_df["Brand Name_y"]
    return new_df
    
        
def transform_sat_kpi(df, df1, df2):
    rows = []
    f_col = df1.iloc[5:].ffill()
    s_col = df2.iloc[5:]
    
    for col in df:
        label = df[col][3]
        temp_df = pd.DataFrame({
            "Question": f_col,
            "Met": s_col,
            "Brand Name": label
        })
        if df[col][4] == "Rank":
            temp_df["Rank"] = df[col][5:]
        else:
            temp_df["%"] = df[col][5:]

        rows.append(temp_df)
    return merge_transform_sat_kpi(rows)
```

```python
def transform_brand_rank_kpis(df):
    rows = []

    f_col = df["Table of Contents"][5:].ffill()
    s_col = df["Unnamed: 1"][5:]

    for col in df:
        temp_df = pd.DataFrame({
            "Question": f_col,
            "Met": s_col,
        })

        if df[col].iloc[4] == "%":
            temp_df["%"] = df[col].iloc[5:]
        elif df[col].iloc[4] == "Rank":
            temp_df["Rank"] = df[col].iloc[5:]
        temp_df["Brand Name"] =  df[col].iloc[3]
        rows.append(temp_df)
    return rows
```

```python
def merge_brk(df_list):
    merged_list = []
    d1 = []
    d2 = []
    df_list = df_list[2:]
    for i in range(len(df_list)):
        if i % 2 == 0:
            d1.append(df_list[i])
        else:
            d2.append(df_list[i])
    for ldf, rdf in zip(d1, d2):
        merged = pd.merge(ldf, rdf, on=["Question", "Met"], how="inner")
        merged_list.append(merged)
    new_df = pd.concat(merged_list, axis=0, ignore_index=True)
    new_df = new_df.drop(columns=["Brand Name_y"])
    new_df = new_df.rename(columns={"Brand Name_x": "Brand Name"})

    reorder = ["Question", "Met", "%", "Rank", "Brand Name"]
    new_df = new_df.reindex(columns=reorder)
    return new_df

```

```python
def transform_hist_met_kpi(df, df2, df3):
    rows = []
    for col in df3:
        label = df3[col][3]
        temp_df = pd.DataFrame({
            "Question": df[3:].ffill(),
            "Met": df2[3:],
            "Quarter": label
        })

        if df3[col].iloc[2] == "%":
            temp_df["%"] = df3[col].iloc[2:]
        elif df3[col].iloc[2] == "Rank":
            temp_df["Rank"] = df3[col].iloc[2:]
        rows.append(temp_df)
    return rows
```

```python
def merge_hmk(df_list):
    d1 = []
    d2 = []
    merged_l = []
    for i in range(len(df_list)):
        if i % 2 == 0:
            d1.append(df_list[i])
        else:
            d2.append(df_list[i])
    
    for ld, rd in zip(d1, d2):
        merged = pd.merge(ld, rd, on=["Question", "Met"])
        merged_l.append(merged)
    new_df = pd.concat(merged_l, ignore_index=True)
    new_df = new_df.drop(columns=["Quarter_y"])
    new_df = new_df.rename(columns={"Quarter_x": "Quarter"})
    return new_df
```

```python
def transform_kpi_index(df):
    df = df.reset_index(drop=True)
    rows = []
    o_col = df.iloc[:,2:]
    for col in o_col:
        m_idx = o_col[col].first_valid_index()
        if o_col[col].iloc[m_idx+1] == "Mean":    
            # Locating the cells to copy over
            f_col = df["Table of Contents"].iloc[m_idx+2:].ffill()
            s_col = df["Unnamed: 1"].iloc[m_idx+2:]
            m_col = o_col[col].iloc[m_idx+2:]
            u_col = df["Unnamed: 3"].iloc[m_idx+2:]
            d_col = o_col[col].iloc[m_idx]
            # creating the tables
            temp_df = pd.DataFrame({
                "Question": f_col,
                "Met": s_col,
                "Mean": m_col,
                "Unweighted base": u_col,
                "Dimension": d_col
            })   
        rows.append(temp_df)
    return rows 
```

```python
def process_merged_dfs(df, names, op):
    merge_kpis = []
    for name in names:
        result = op(df[name])
        if isinstance(result, list):
            merge_kpis.extend(result)
        else:
            merge_kpis.append(result)
    new_df = pd.concat(merge_kpis, axis=0)
    return new_df
```

```python
merged_kpis_all = process_merged_dfs(output, names_kpi, transform_kpi_index)
```

```python
def transform_cost_historicals(df):
    df = df.reset_index(drop=True)
    rows = []
    o_col = df.iloc[:,2:]
    m_idx = o_col["Unnamed: 2"].first_valid_index()

    temp_df = pd.DataFrame({
        "Question": df["Table of Contents"].iloc[m_idx+3:].ffill(),
        "Met": df["Unnamed: 1"].iloc[m_idx+3:],
        df["Unnamed: 2"].iloc[m_idx+1]: df["Unnamed: 2"].iloc[m_idx+3:],
        df["Unnamed: 3"].iloc[m_idx+1]: df["Unnamed: 3"].iloc[m_idx+3:],
        df["Unnamed: 4"].iloc[m_idx+1]: df["Unnamed: 4"].iloc[m_idx+3:],
        # Add the dimension
        "Dimension": df["Table of Contents"].iloc[0]
    })
    rows.append(temp_df)
    return rows 
```

```python
def process_merged_dfs(df, names, op):
    merge_kpis = []
    for name in names:
        result = op(df[name])
        if isinstance(result, list):
            merge_kpis.extend(result)
        else:
            merge_kpis.append(result)
    new_df = pd.concat(merge_kpis, axis=0)
    return new_df
```
